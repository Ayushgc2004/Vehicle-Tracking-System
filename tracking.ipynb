```python
import cv2
from ultralytics import YOLO
import numpy as np
import matplotlib.pyplot as plt
from difflib import SequenceMatcher
import re
from paddleocr import PaddleOCR

# Load the YOLO models for color detection and license plate detection
color_model = YOLO('/content/color_detection3.pt')
plate_model = YOLO('/content/license_plate_detection.pt')

# Initialize PaddleOCR model for license plate recognition
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # You can change 'en' to any supported language

# Function to clean the plate number by removing any symbols (only keep alphanumeric characters)
def clean_plate_number(plate_number):
    return re.sub(r'[^A-Za-z0-9]', '', plate_number)

# Function to generate all possible substrings of the input string with length >= 4
def get_substrings(plate_number):
    substrings = set()
    length = len(plate_number)
    for i in range(length):
        for j in range(i+1, length+1):
            substring = plate_number[i:j]
            if len(substring) >= 4:  # Only include substrings with length >= 4
                substrings.add(substring)
    return substrings

# Function to calculate the match confidence percentage between two strings
def match_confidence(str1, str2):
    return SequenceMatcher(None, str1, str2).ratio() * 100

# Function to process the image and return vehicles with the target color
def process_image_for_color(image, target_color):
    results = color_model(image)
    detected_boxes = []

    for result in results[0].boxes.data:
        color_label = results[0].names[int(result[5])]  # Extract the color label from the result
        if color_label == target_color:
            # Get the bounding box coordinates (x1, y1, x2, y2)
            x1, y1, x2, y2 = map(int, result[:4])
            detected_boxes.append((x1, y1, x2, y2))

    return detected_boxes  # Return bounding boxes of vehicles with the target color

# Function to process the image and recognize license plates in the target vehicles
def process_image_for_license_plate(image, detected_boxes, target_plate_number):
    detected_frame = False
    cleaned_target_plate = clean_plate_number(target_plate_number)  # Clean the target plate number
    substrings = get_substrings(cleaned_target_plate)  # Get substrings from the cleaned plate number

    for box in detected_boxes:
        x1, y1, x2, y2 = box

        # Extract the region of interest (ROI) - the detected vehicle
        license_plate_roi = image[y1:y2, x1:x2]

        # Use PaddleOCR to recognize text in the ROI
        ocr_result = ocr.ocr(license_plate_roi, rec=True, cls=True)

        # Debugging: Print the OCR result structure
        print(f"OCR Result for box {box}: {ocr_result}")

        # Check if ocr_result is not None and contains valid results
        if ocr_result and isinstance(ocr_result, list) and len(ocr_result) > 0:
            for result in ocr_result:
                if result and isinstance(result, list) and len(result) > 0 and isinstance(result[0], list) and len(result[0]) > 1:
                    recognized_text = result[0][1][0]  # Get the recognized text
                    confidence = result[0][1][1]       # Get the confidence score
                    cleaned_recognized_text = clean_plate_number(recognized_text)  # Clean the recognized text
                    print(f"Detected License Plate: {cleaned_recognized_text} with confidence {confidence:.2f}%")

                    # Check if the recognized text matches any of the substrings with 85% or higher confidence
                    for substring in substrings:
                        confidence_match = match_confidence(cleaned_recognized_text, substring)
                        if confidence_match >= 85:
                            # Draw the bounding box and recognized text on the image
                            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
                            cv2.putText(image, cleaned_recognized_text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                                        0.9, (255, 0, 0), 2, cv2.LINE_AA)
                            print(f"Match found with confidence: {confidence_match:.2f}% for substring: {substring}")
                            detected_frame = True  # Mark this frame as one to be saved/displayed
                            break  # Stop checking substrings after a match is found
                else:
                    print(f"Unexpected OCR result structure for box {box}: {ocr_result}")
        else:
            print(f"No text detected in box {box} or invalid OCR result")

    return image if detected_frame else None

# Function to process video input and apply the combined detection pipeline
def process_video_input(video_path, target_color, target_plate_number, output_path):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error: Unable to open video file {video_path}")
        return

    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Output video writer (to save the filtered frames)
    out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

    frame_count = 0  # Counter for frames
    target_frame_found = False  # Flag to indicate if the target frame is found

    while True:
        ret, frame = cap.read()
        if not ret:
            print("End of video stream or cannot fetch the frame.")
            break

        frame_count += 1  # Increment frame counter

        # Step 1: Detect vehicles of the target color
        detected_boxes = process_image_for_color(frame, target_color)
        print(f"Frame {frame_count}: Detected {len(detected_boxes)} vehicle(s) with color {target_color}")

        if detected_boxes:
            # Step 2: Process these vehicles for license plate detection and OCR
            annotated_frame = process_image_for_license_plate(frame, detected_boxes, target_plate_number)

            # Step 3: If a matching frame is found, display and write it to output, then terminate the process
            if annotated_frame is not None:
                # Display the frame with the detected license plate
                plt.figure(figsize=(10, 6))
                plt.imshow(cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB))
                plt.title(f"Frame {frame_count} - Color: {target_color}")
                plt.axis('off')  # Hide axis
                plt.show()

                # Write the frame to the output video
                out.write(annotated_frame)

                # Print that the target frame was found and terminate the process
                print(f"Target vehicle with color '{target_color}' and plate number detected in frame {frame_count}")
                target_frame_found = True
                break  # Exit the while loop

    cap.release()
    out.release()

    if target_frame_found:
        print(f"Processing complete. Output saved to {output_path}")
    else:
        print("No matching vehicle found in the video.")


# Example usage for video input
video_path = '/content/color_test_video.mp4'
target_color = 'white'  # Specify the target vehicle color
target_plate_number = 'K-884-RS'  # Provide the full plate number
output_path = '/content/output_filtered_with_license_plate.mp4'

process_video_input(video_path, target_color, target_plate_number, output_path)
```

```python
import cv2
from ultralytics import YOLO
import numpy as np
import matplotlib.pyplot as plt
from difflib import SequenceMatcher
import re
from paddleocr import PaddleOCR
from datetime import timedelta, datetime

# Load the YOLO models for color detection and license plate detection
color_model = YOLO('/content/color_detection3.pt')
plate_model = YOLO('/content/license_plate_detection.pt')

# Initialize PaddleOCR model for license plate recognition
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # You can change 'en' to any supported language

# Function to clean the plate number by removing any symbols (only keep alphanumeric characters)
def clean_plate_number(plate_number):
    return re.sub(r'[^A-Za-z0-9]', '', plate_number)

# Function to generate all possible substrings of the input string with length >= 4
def get_substrings(plate_number):
    substrings = set()
    length = len(plate_number)
    for i in range(length):
        for j in range(i+1, length+1):
            substring = plate_number[i:j]
            if len(substring) >= 4:  # Only include substrings with length >= 4
                substrings.add(substring)
    return substrings

# Function to calculate the match confidence percentage between two strings
def match_confidence(str1, str2):
    return SequenceMatcher(None, str1, str2).ratio() * 100

# Function to process the image and return vehicles with the target color
def process_image_for_color(image, target_color):
    results = color_model(image)
    detected_boxes = []

    for result in results[0].boxes.data:
        color_label = results[0].names[int(result[5])]  # Extract the color label from the result
        if color_label == target_color:
            # Get the bounding box coordinates (x1, y1, x2, y2)
            x1, y1, x2, y2 = map(int, result[:4])
            detected_boxes.append((x1, y1, x2, y2))

    return detected_boxes  # Return bounding boxes of vehicles with the target color

# Function to process the image and recognize license plates in the target vehicles
def process_image_for_license_plate(image, detected_boxes, target_plate_number):
    detected_frame = False
    cleaned_target_plate = clean_plate_number(target_plate_number)  # Clean the target plate number
    substrings = get_substrings(cleaned_target_plate)  # Get substrings from the cleaned plate number

    for box in detected_boxes:
        x1, y1, x2, y2 = box

        # Extract the region of interest (ROI) - the detected vehicle
        license_plate_roi = image[y1:y2, x1:x2]

        # Use PaddleOCR to recognize text in the ROI
        ocr_result = ocr.ocr(license_plate_roi, rec=True, cls=True)

        # Debugging: Print the OCR result structure
        print(f"OCR Result for box {box}: {ocr_result}")

        # Check if ocr_result is not None and contains valid results
        if ocr_result and isinstance(ocr_result, list) and len(ocr_result) > 0:
            for result in ocr_result:
                if result and isinstance(result, list) and len(result) > 0 and isinstance(result[0], list) and len(result[0]) > 1:
                    recognized_text = result[0][1][0]  # Get the recognized text
                    confidence = result[0][1][1]       # Get the confidence score
                    cleaned_recognized_text = clean_plate_number(recognized_text)  # Clean the recognized text
                    print(f"Detected License Plate: {cleaned_recognized_text} with confidence {confidence:.2f}%")

                    # Check if the recognized text matches any of the substrings with 85% or higher confidence
                    for substring in substrings:
                        confidence_match = match_confidence(cleaned_recognized_text, substring)
                        if confidence_match >= 85:
                            # Draw the bounding box and recognized text on the image
                            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
                            cv2.putText(image, cleaned_recognized_text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                                        0.9, (255, 0, 0), 2, cv2.LINE_AA)
                            print(f"Match found with confidence: {confidence_match:.2f}% for substring: {substring}")
                            detected_frame = True  # Mark this frame as one to be saved/displayed
                            break  # Stop checking substrings after a match is found
                else:
                    print(f"Unexpected OCR result structure for box {box}: {ocr_result}")
        else:
            print(f"No text detected in box {box} or invalid OCR result")

    return image if detected_frame else None

# Function to convert the frame count to the corresponding timestamp
def get_frame_timestamp(start_time, fps, frame_count):
    return start_time + timedelta(seconds=frame_count / fps)

# Function to process video input and apply the combined detection pipeline
def process_video_input(video_path, target_color, target_plate_number, output_path, start_time_str):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error: Unable to open video file {video_path}")
        return

    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Convert start_time_str to a datetime object
    start_time = datetime.strptime(start_time_str, '%H:%M:%S')

    # Output video writer (to save the filtered frames)
    out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

    frame_count = 0  # Counter for frames
    target_frame_found = False  # Flag to indicate if the target frame is found
    found_time = None  # Variable to store the timestamp of detection

    while True:
        ret, frame = cap.read()
        if not ret:
            print("End of video stream or cannot fetch the frame.")
            break

        frame_count += 1  # Increment frame counter

        # Step 1: Detect vehicles of the target color
        detected_boxes = process_image_for_color(frame, target_color)
        print(f"Frame {frame_count}: Detected {len(detected_boxes)} vehicle(s) with color {target_color}")

        if detected_boxes:
            # Step 2: Process these vehicles for license plate detection and OCR
            annotated_frame = process_image_for_license_plate(frame, detected_boxes, target_plate_number)

            # Step 3: If a matching frame is found, display and write it to output, then terminate the process
            if annotated_frame is not None:
                # Display the frame with the detected license plate
                plt.figure(figsize=(10, 6))
                plt.imshow(cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB))
                plt.title(f"Frame {frame_count} - Color: {target_color}")
                plt.axis('off')  # Hide axis
                plt.show()

                # Write the frame to the output video
                out.write(annotated_frame)

                # Calculate the timestamp for the detected frame
                found_time = get_frame_timestamp(start_time, fps, frame_count)
                print(f"Target vehicle with color '{target_color}' and plate number detected at {found_time.strftime('%H:%M:%S')}")
                target_frame_found = True
                break  # Exit the while loop

    cap.release()
    out.release()

    if target_frame_found:
        print(f"Processing complete. Output saved to {output_path}")
        return found_time.strftime('%H:%M:%S')  # Return the detection timestamp
    else:
        print("No matching vehicle found in the video.")
        return None

# Function to handle multiple videos, collecting all inputs at once
def process_multiple_videos():
    num_videos = int(input("Enter the number of videos you want to process: "))
    target_color = input("Enter the target vehicle color: ")
    target_plate_number = input("Enter the target license plate number: ")

    # Collect all video paths and start times at once
    video_details = []
    for i in range(num_videos):
        video_path = input(f"Enter the path for video {i+1}: ")
        start_time = input(f"Enter the start time of video {i+1} (HH:MM:SS): ")
        video_details.append((video_path, start_time))

    # List to store the results of each video
    results = []

    # Process each video after collecting all inputs
    for i, (video_path, start_time) in enumerate(video_details):
        output_path = f'/content/output_filtered_video_{i+1}.mp4'

        # Process each video and get the detection time (if any)
        detection_time = process_video_input(video_path, target_color, target_plate_number, output_path, start_time)

        # Save the result for this video
        if detection_time:
            results.append(f"Vehicle found in video {i+1} at time {detection_time}")
        else:
            results.append(f"Vehicle not found in video {i+1}")

    # Printing the final results
    print("\nSummary of Results:")
    for result in results:
        print(result)

process_multiple_videos()
```
